```python
import numpy as np
import pandas as pd
import re
```

```python
importing = open('data/day-3.txt', 'r').read()

muls_and_idx = []

# Regex pattern
pattern = r"mul\(\d+,\d+\)"

# Find matches with their indices
matches = [(match.group(), match.start(), match.end()) for match in re.finditer(pattern, importing)]
muls = []

# Print results
for match_text, start_idx, end_idx in matches:
    muls_and_idx.append((match_text, start_idx))
    muls.append(match_text)

muls
```

```python
dff = pd.DataFrame(muls_and_idx)
```

```python
neww = []
integers = []

for i in range(len(muls)):
    n = muls[i].replace('mul','').replace('(','').replace(')','').replace(',', ' ')
    neww.append(n.split())
    integers.append((int(neww[i][0]), int(neww[i][1])))

df = pd.DataFrame(integers)
result_part_1 = np.sum(df[0]*df[1])
```

```python
importing = open('data/day-3.txt', 'r').read()

pattern_does = r"\bdo\(\)|\bdon't\(\)"
matches_does = re.findall(pattern_does, importing)
matches_does
```

```python
importing = open('data/day-3.txt', 'r').read()

list_donts = []

for _ in range(27):
    num_do = importing.find("don't()")
    num_mul = importing[num_do:].find('mul(')
    importing = importing[num_do:][num_mul:]

    list_donts.append((num_do, re.findall(pattern, importing)[0]))
```

```python
importing = open('data/day-3.txt', 'r').read()
list_does = []

for _ in range(45):
    num_do = importing.find('do()')
    num_mul = importing[num_do:].find('mul(')
    importing = importing[num_do:][num_mul:]

    list_does.append((num_do,re.findall(pattern, importing)[0]))
```

```python
list_does
```

```python
test_df = pd.DataFrame(list_donts)

for i in range(len(test_df)-1):
    test_df.loc[0, 'indices_start'] = test_df.loc[0,0]
    test_df.loc[i+1, 'indices_start'] = test_df.loc[i, 'indices_start'] + test_df.loc[i+1,0]

test_df = test_df.drop(columns=0).rename(columns={1:'to_multiply'})
test_df['do_or_dont'] = 'N'
```

```python
test_does = pd.DataFrame(list_does)

for i in range(len(test_does)-1):
    test_does.loc[0, 'indices_start'] = test_does.loc[0,0]
    test_does.loc[i+1, 'indices_start'] = test_does.loc[i, 'indices_start'] + test_does.loc[i+1,0]

test_does = test_does.drop(columns=0).rename(columns={1:'to_multiply'})
test_does['do_or_dont'] = 'Y'

df_does_and_donts = pd.concat([test_df, test_does]).sort_values(by='indices_start')
```

```python
dff.rename(columns={0:'to_multiply',
                    1:'indices_start'}, inplace=True)

ok = pd.merge(dff, df_does_and_donts, on=['to_multiply','indices_start'], how='outer') #['do_or_dont'].value_counts()
ok = ok.sort_values(by='indices_start')
ok.head(30)
```

```python
# modifying the np.nans so that we know if we have to apply the function multiply

ok.loc[0, 'do_or_dont']='Y'

for i in range(len(ok)-1):
    if pd.isna(ok.loc[i+1,'do_or_dont']):
        if ok.loc[i, 'do_or_dont']=='Y':
            ok.loc[i+1,'do_or_dont']='Y'
        elif ok.loc[i,'do_or_dont']=='N':
            ok.loc[i+1,'do_or_dont']='N'

# equivalent to
# ok['do_or_dont'].fillna("ffill", inplace=True)
```
