```python
def finding_answer(df):
    possibilities = []
    length = len(df.iloc[0,:].dropna())-1
    for _ in range(length*length*10):
        for i in range(length):
            for _ in range(5):
                ops = (add, mul)
                operation = random.choice(ops)
                answer_1 = operation(df.iloc[i,i], df.iloc[i,i+1])
                for _ in range(5):
                    ops = (add, mul)
                    operation = random.choice(ops)
                    answer = operation(answer_1, df.iloc[i,i+1])
                # print('i am here, at i =', i)
                if answer not in possibilities and np.isnan(answer)==False:
                    possibilities.append(answer)

    return possibilities
```
